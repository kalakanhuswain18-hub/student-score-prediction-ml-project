```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error, mean_squared_error , r2_score
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
data=pd.read_csv("student_performance.csv")
data
```

```python
x=data[["weekly_self_study_hours"]]
y=data["total_score"]
model=LinearRegression()
model.fit(x,y)
pred_score=model.predict(x)
mae=mean_absolute_error(y,pred_score)
mse=mean_squared_error(y,pred_score)
rmse=np.sqrt(mse)
r2=r2_score(y,pred_score)
print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"R^2 Score: {r2:.3f}")
#Histplot
plt.figure(figsize=(10,8))
plt.hist(data["total_score"] ,bins=30 , color="green" , edgecolor="black")
plt.title("Distribution of final exam score.....")
plt.xlabel("Final exam score ")
plt.ylabel("Number of student")
plt.grid(True)
plt.show()
```

```python
#scatter plot 
plt.figure(figsize=(10,8))
plt.scatter(x,y, color="green" , label="Actual Score")
plt.plot(x,pred_score , color="red" , label="predicted score regression line")
plt.title("Model prediction vs actual score.....")
plt.xlabel("Study hours per week ")
plt.ylabel("Final output ")
plt.grid(True)
plt.show()
new_hours=9
predicted_score=model.predict(pd.DataFrame([[new_hours]] ,columns=["weekly_self_study_hours"]))
print(f"predicted final score for {new_hours} hours is {predicted_score}")
```
